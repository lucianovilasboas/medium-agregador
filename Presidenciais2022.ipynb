```python
import pandas as pd
from matplotlib import pyplot as plt
import plotly.express as px
from google.colab import drive
drive.mount('drive')

!ls "drive/My Drive/Python_scripts/Eleicoes2022/Poder360"
```

```python
# define o caminho do aquivo com os dados
data_path = "drive/My Drive/Python_scripts/Eleicoes2022/Poder360/consulta92435963_from_01_01_2019_to_18_02_2022.csv"
df = pd.read_csv(data_path, sep=';')

# converte a coluna 'data_pesquisa' para 'datetime64[ns]'  
df['data_pesquisa']= df['data_pesquisa'].astype('datetime64[ns]')
```

```python
df.head()
```

```python
# definindo alguns filtros
candidatos = ["Lula","Bolsonaro","Ciro","Sergio Moro"]

# definindo as cores para os candidatos 
cores = ['red', 'green', 'blue', 'orange']
color_discrete_map = dict(zip(candidatos,cores))

cargo = "Presidente"
tipos  = ["Espontânea", "Estimulada"]
tipo = " e ".join(tipos)

# filtrando os dados por turno (1° e 2°)/cargo (Presidente)/tipo(Estimulada)
df_primeiro_turno = df[df['candidato'].isin(candidatos) & df['tipo'].isin(tipos) ].query('turno == 1 and cargos_id == 3')
df_segundo_turno  = df[df['candidato'].isin(candidatos) & df['tipo'].isin(tipos)].query('turno == 2 and cargos_id == 3')
```

```python
# primeira visualização 
fig = px.scatter(df_primeiro_turno,\
   title=f"Meu Agregador - {cargo} 2022 - 1° Turno ({tipo})",\
   x="data_pesquisa", y="percentual",  color="candidato",\
   color_discrete_map=color_discrete_map,\
   hover_name="instituto", hover_data={'tipo': True} 
)
fig.show()

# primeira visualização
fig = px.scatter(df_segundo_turno,\
   title=f"Meu Agregador - {cargo} 2022 - 2° Turno ({tipo})",\
   x="data_pesquisa", y="percentual", color="candidato",\
   color_discrete_map=color_discrete_map,\
   template="simple_white",  hover_name="instituto",\
   hover_data={'tipo': True}
)
fig.show()
```

```python
fig = px.scatter(df_primeiro_turno,title=f"Meu Agregador - {cargo} 2022 - 1° Turno ({tipo})",\
    x="data_pesquisa", y="percentual",  color="candidato", color_discrete_map=color_discrete_map,\
    hover_name="instituto",\
    trendline="rolling", trendline_options=dict(function="mean", window=10))

fig.update_layout(hovermode="x")
fig.show()


fig = px.scatter(df_segundo_turno,title=f"Meu Agregador - {cargo} 2022 - 2° Turno ({tipo})",\
    x="data_pesquisa", y="percentual",  color="candidato", color_discrete_map=color_discrete_map,\
    hover_name="instituto",\
    trendline="rolling", trendline_options=dict(function="mean", window=10))
fig.update_layout(hovermode="x")
fig.show()
```
